```python
from pathlib import Path
import os
import subprocess

workdir = Path('/workspaces/post-it-core/web')
os.chdir(workdir)

files = [
    Path('public/icons/post-this-192.png'),
    Path('public/icons/post-this-512.png'),
    Path('public/icons/post-this-1024.png'),
]

proc = subprocess.run(
    ['node', 'scripts/generate-icons.mjs'],
    capture_output=True,
    text=True,
    cwd=str(workdir),
)

combined_output = (proc.stdout or '') + (proc.stderr or '')

png_magic = b'\x89PNG\r\n\x1a\n'
results = []
for p in files:
    exists = p.exists()
    is_png = False
    size_bytes = None
    if exists:
        with p.open('rb') as f:
            sig = f.read(8)
        is_png = sig == png_magic
        size_bytes = p.stat().st_size
    results.append({
        'path': str(p),
        'exists': exists,
        'is_png': is_png,
        'size_bytes': size_bytes,
    })

print('1) command output')
print(combined_output if combined_output.strip() else '(no output)')
print('2) bevestiging bestanden')
for r in results:
    print(f"{r['path']}: exists={r['exists']}, png={r['is_png']}, size_bytes={r['size_bytes']}")
print(f"returncode={proc.returncode}")
```
